# **Task 2 â€“ Spatial Averaging**

```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
import os

# Define kernel sizes
kernel_sizes = [3, 10, 20]

# Load the original color image
img = cv2.imread('images/input.jpg')

# Check if image is loaded
if img is None:
    raise FileNotFoundError("Image not found. Please check the path: 'images/input.jpg'")

# Create output directory if it doesn't exist
os.makedirs('outputs/task2', exist_ok=True)

# Store results (first image is original)
results = [("Original", img)]

# Apply blurring and store
for k in kernel_sizes:
    blurred = cv2.blur(img, (k, k))
    results.append((f"{k}x{k}", blurred))
    
    # Save output
    cv2.imwrite(f'outputs/task2/blur_{k}x{k}.jpg', blurred)

# Plot all images in one row
plt.figure(figsize=(18, 5))
for i, (label, image) in enumerate(results):
    plt.subplot(1, len(results), i + 1)
    plt.imshow(cv2.cvtColor(image, cv2.COLOR_BGR2RGB))
    plt.title(f"{label} Blur" if label != "Original" else "Original Image")
    plt.axis('off')

plt.tight_layout()
plt.show()
```
