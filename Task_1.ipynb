# **Task 1 - Intensity Level Reduction**

```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
import os

# Load the original color image 
img = cv2.imread('images/input.jpg')

# Check if image loaded
if img is None:
    raise FileNotFoundError("Image not found. Check path: 'images/input.png'")

# Set intensity levels
levels = 4
factor = 256 // levels  # Correct factor: 64

# Reduce intensity on each channel
reduced = (img // factor) * factor
reduced = reduced.astype(np.uint8)  # Ensure proper type

# Create output directory
os.makedirs('outputs/task1', exist_ok=True)
cv2.imwrite(f'outputs/task1/reduced_{levels}_levels.jpg', reduced)

# Display both images
plt.figure(figsize=(10, 5))

plt.subplot(1, 2, 1)
plt.imshow(cv2.cvtColor(img, cv2.COLOR_BGR2RGB))
plt.title("Original Image")
plt.axis('off')

plt.subplot(1, 2, 2)
plt.imshow(cv2.cvtColor(reduced, cv2.COLOR_BGR2RGB))
plt.title(f"Reduced to {levels} Levels")
plt.axis('off')

plt.tight_layout()
plt.show()
```
