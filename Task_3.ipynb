# **Task 3 – Rotation by 45° and 90°**

```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
import os

def rotate_image(image, angle):
    h, w = image.shape[:2]
    center = (w // 2, h // 2)
    M = cv2.getRotationMatrix2D(center, angle, 1.0)
    rotated = cv2.warpAffine(image, M, (w, h))
    return rotated

# Load original image
img = cv2.imread('images/input.jpg')

# Ensure output folder exists
os.makedirs('outputs/task3', exist_ok=True)

# Initialize results with original image first
results = [("Original", img)]

# Angles to rotate
angles = [45, 90]

for angle in [45, 90]:
    rotated = rotate_image(img, angle)
    results.append((f"Rotated {angle}°", rotated))
    cv2.imwrite(f'outputs/task3/rotated_{angle}.jpg', rotated)

# Display all images in one row
plt.figure(figsize=(15, 5))
for i, (label, image) in enumerate(results):
    plt.subplot(1, len(results), i + 1)
    plt.imshow(cv2.cvtColor(image, cv2.COLOR_BGR2RGB))
    plt.title(label)
    plt.axis('off')

plt.tight_layout()
plt.show()
```
